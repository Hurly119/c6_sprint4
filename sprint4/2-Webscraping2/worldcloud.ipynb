```python
from bs4 import BeautifulSoup
import requests
```

```python
with open("your_posts_1.html",encoding="utf8") as page:
    soup = BeautifulSoup(page,"html.parser")
```

```python
contents = soup.find_all("div",class_="_2pin")
```

```python
contents = [i.get_text() for i in contents]
```

```python
contents
```

```python
word_cloud = " ".join(contents)
```

```python
import matplotlib.pyplot as plt
from wordcloud import WordCloud

wordcloud = WordCloud()

wordcloud.generate(word_cloud)

plt.imshow(wordcloud,interpolation="bilinear")
plt.axis("off")
```

```python
punctuations = '''!()-[]{};:'"\,<>./?@#$%^&*_~‚Äú‚Äù‚Äô‚Äî'''
uninteresting_words = ["the", "a", "to", "if", "is", "it", "of", "and", "or", "an", "as", "i", "me", "my", \
    "we", "our", "ours", "you", "your", "yours", "he", "she", "him", "his", "her", "hers", "its", "they", "them", \
    "their", "what", "which", "who", "whom", "this", "that", "am", "are", "was", "were", "be", "been", "being", \
    "have", "has", "had", "do", "does", "did", "but", "at", "by", "with", "from", "here", "when", "where", "how", \
    "all", "any", "both", "each", "few", "more", "some", "such", "no", "nor", "too", "very", "can", "will", "just"]
    # LEARNER CODE START HERE
frequencies = dict()
for punctuation in punctuations:
    if punctuation in word_cloud:
        file_contents = word_cloud.replace(punctuation, "")
    
good_words = " ".join([word for word in file_contents.lower().split() if word not in uninteresting_words])

wordcloud.generate(good_words)

plt.imshow(wordcloud,interpolation="bilinear")
plt.axis("off")
```
